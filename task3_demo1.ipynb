```python
import numpy as np
import matplotlib.pyplot as plt

def dragon_curve_sequence(n):
    seq = "F"
    rules = {"F": "F+G", "G": "F-G"}
    for _ in range(n):
        seq = "".join(rules.get(c, c) for c in seq)
    return seq

def draw_dragon_curve(seq, step=5):
    x, y = 0, 0
    angle = 0
    X, Y = [x], [y]

    for c in seq:
        if c in ["F", "G"]:
            x += step * np.cos(np.radians(angle))
            y += step * np.sin(np.radians(angle))
            X.append(x)
            Y.append(y)
        elif c == "+":
            angle += 90
        elif c == "-":
            angle -= 90
    return X, Y

seq = dragon_curve_sequence(15)
X, Y = draw_dragon_curve(seq, step=5)


plt.figure(figsize=(8,8))
plt.plot(X, Y, color="blue")
plt.axis("equal")
plt.title("Dragon Curve (Iteration 15)")
plt.show()
```

```python
plt.figure(figsize=(8,8))
points = len(X)
plt.scatter(X, Y, c=np.linspace(0, 1, points), cmap="viridis", s=1)
plt.axis("equal")
plt.title("Dragon Curve (coloured by progression)")
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.plot(X, Y, color="red", linewidth=0.8)
plt.axis("equal")
plt.title("Dragon Curve (red version)")
plt.show()

plt.figure(figsize=(8,8))
plt.plot(X, Y, color="green", linestyle="dashed")
plt.axis("equal")
plt.title("Dragon Curve (dashed line)")
plt.show()
```
